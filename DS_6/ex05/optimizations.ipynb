## iterations: in all the following subtasks, you need to calculate fines/refund*year for each row and create a new column with the calculated data and measure the time using the magic command %%timeit in the cell
- loop: write a function that iterates through the dataframe using for i in range(0, len(df)), iloc and append() to a list, assign the result of the func- tion to a new column in the dataframe
- do it using iterrows()
- do it using apply() and lambda function
- do it using Series objects from the dataframe
- do it as in the previous subtask but with the method .values

```python
%%timeit -n 1 -r 1

calculate_loop = [
    fines_df.iloc[row_index]['Fines'] / fines_df.iloc[row_index]['Refund'] * fines_df.iloc[row_index]['Year']
    for row_index in range(len(fines_df))
]
fines_df['Calculate'] = calculate_loop
```

```python
fines_df
```

```python
%%timeit -n 1 -r 1

calculate_iterrows = [
    row['Fines'] / row['Refund'] * row['Year']
    for index, row in fines_df.iterrows()
]

fines_df['Calculate'] = calculate_iterrows
```

```python
fines_df
```

```python
%%timeit -n 1 -r 1

fines_df['Calculate'] = fines_df.apply(lambda x: (x['Fines'] / x['Refund']) * x['Year'], axis=1)
```

```python
fines_df
```

```python
%%timeit -n 1 -r 1

fines_df['Calculate'] = fines_df['Fines'] / fines_df['Refund'] * fines_df['Year']
```

```python
%%timeit -n 1 -r 1

fines_df['Calculate'] = fines_df['Fines'].values / fines_df['Refund'].values * fines_df['Year'].values
```

## downcasting
- run df.info(memory_usage=’deep’), pay attention to the Dtype and the mem- ory usage
- make a copy() of your initial dataframe into another dataframe optimized

```python
fines_df.info(memory_usage='deep')
```

```python
optimized_df = fines_df.copy()
optimized_df.reset_index(inplace=True)

fl32_column = ['Fines', 'Calculate']
category_column = ['Make', 'Model', 'CarNumber']

optimized_df['Refund'] = optimized_df['Refund'].astype('int8')

optimized_df[fl32_column] = optimized_df[fl32_column].astype('float32')

optimized_df['Year'] = optimized_df['Year'].astype('int16')

optimized_df[category_column] = optimized_df[category_column].astype('category')
```
