# Exercise 05 : Boxplot

```python
df = pd.read_csv('data/ab-test.csv')
df.drop(columns=['uid'], inplace=True)
df
```

```python
before_df = df[df['time'] == 'before'].drop(columns=['time'])
before_df
```

```python
after_df = df[df['time'] == 'after'].drop(columns=['time'])
after_df
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))

ax2.boxplot([before_df[before_df['group'] == 'control']['diff'], 
             before_df[before_df['group'] == 'test']['diff']],
             tick_labels=['test', 'control'],
            boxprops=dict(color='green', linewidth=3),
            whiskerprops=dict(color='red'), 
            capprops=dict(color='red'),
            medianprops=dict(color='red', linewidth=2)
           )

ax1.boxplot([after_df[after_df['group'] == 'control']['diff'], 
             after_df[after_df['group'] == 'test']['diff']],
             tick_labels=['test', 'control'],
            boxprops=dict(color='green', linewidth=3),
            whiskerprops=dict(color='red'), 
            capprops=dict(color='red'),
            medianprops=dict(color='red', linewidth=2)
           )

ax1.set_title('After')
ax2.set_title('Before')
ax1.grid(True)
ax2.grid(True)

ax1.set_ylim(-340, 40)
ax2.set_ylim(-340, 40)

fig.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)

plt.show()
```
