```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
#  
X_original = data[feature_cols].copy() #  catboost
X = data[feature_cols].copy()

#     2   (1, 0)
X['Sex'] = X['Sex'].map({'Female': 0, 'Male': 1})
X['HasPartner'] = X['HasPartner'].map({'No': 0, 'Yes': 1})
X['HasChild'] = X['HasChild'].map({'No': 0, 'Yes': 1})
X['HasPhoneService'] = X['HasPhoneService'].map({'No': 0, 'Yes': 1})
```

```python
unprocessed_cat_features = [
    'HasMultiplePhoneNumbers',
    'HasInternetService',
    'HasOnlineSecurityService',
    'HasOnlineBackup',
    'HasDeviceProtection',
    'HasTechSupportAccess',
    'HasOnlineTV',
    'HasMovieSubscription',
    'HasContractPhone',
    'IsBillingPaperless',
    'PaymentMethod'
]
```

```python
encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(X[unprocessed_cat_features])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(unprocessed_cat_features), index=X.index)
X_encoded = pd.concat([X, one_hot_df], axis=1)

X_encoded = X_encoded.drop(unprocessed_cat_features, axis=1)
```

```python
scaler = StandardScaler()
scaled_nums = scaler.fit_transform(X[['ClientPeriod','MonthlySpending','TotalSpent']])
X_encoded[['ClientPeriod','MonthlySpending','TotalSpent']] = scaled_nums
X_encoded
```

```python
features = [
    'ClientPeriod',
    'MonthlySpending',
    'TotalSpent',
    'Churn',
    'maybe',
    'Sex',
    'IsSeniorCitizen',
    'HasPartner',
    'HasChild',
    'HasPhoneService',
]

cor = X_encoded.copy()
cor['Churn'] = data['Churn']
cor['maybe'] = maybe

corrmat = cor[features].corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);

k = 20
cols = corrmat.nlargest(k, 'Churn')['Churn'].index
cm = np.corrcoef(cor[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
#         
#     MainClient
```

```python
#       
X_encoded['MainClient'] = maybe
```

```python
y = data['Churn']
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, shuffle=True, test_size=0.2, random_state=44)
```

```python
#  GridSearCV   (    )    
log_reg = LogisticRegression(max_iter=300, class_weight='balanced')

param_grid = {'C': [100, 10, 1, 0.1, 0.01, 0.001]}

gridsearch = GridSearchCV(log_reg, param_grid, refit=True, cv=5, scoring='roc_auc')

gridsearch.fit(X_train, y_train)

best_model = gridsearch.best_estimator_

print(" :", gridsearch.best_params_)
print(" AUC   :", gridsearch.best_score_)
```

```python
#    
y_pred_proba = best_model.predict_proba(X_test)[:, 1]

auc_score = roc_auc_score(y_test, y_pred_proba)
print("AUC   :", auc_score)
```

```python
log_reg_cv = LogisticRegressionCV(cv=5, random_state=44, max_iter=200, class_weight='balanced')
log_reg_cv.fit(X_train, y_train)

#    
y_pred_proba_cv = log_reg_cv.predict_proba(X_test)[:, 1]

auc_score_cv = roc_auc_score(y_test, y_pred_proba_cv)

print("  C:", log_reg_cv.C_)
print("AUC   :", auc_score_cv)
```
