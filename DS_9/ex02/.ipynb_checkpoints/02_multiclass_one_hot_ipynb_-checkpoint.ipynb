## 0. Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
```

## 1. Preprocessing

1. Read the file [`checker-submits.csv`](https://drive.google.com/file/d/14voc4fNJZiLEFaZyd8nEG-lQt5JjatYw/view?usp=sharing).
2. Create a dataframe `df` with the columns: `uid`, `labname`, `numTrials`, `hour`, `dayofweek` where `hour` is extracted from the `timestamp` as well as the `dayofweek` (`0` is Monday, `6` is Sunday). We will try to predict the day of the week having data about which user made a commit for which lab at which hour and which try it was.
3. Using `OneHotEncoder()` transform your categorical features, delete from the dataframe the initial columns.
4. Use `StandardScaler()` and scale your continuous features.
5. Save the dataframe as `dayofweek.csv`.
6. Before trying out different algorithms, find out the accuracy of the naive algorithms – the one that predicts everything as the most popular class.

```python
data = pd.read_csv('../data/checker_submits.csv')
data.head()
```

```python
data['timestamp'].info()
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
df = pd.DataFrame()

df = data[['uid', 'labname', 'numTrials']]
df['hour'] = data['timestamp'].dt.hour
df['dayofweek'] = data['timestamp'].dt.dayofweek
df.head()
```

```python
scaler = StandardScaler()

scaled_data = scaler.fit_transform(df[['numTrials', 'hour']])

scaled_df = pd.DataFrame(scaled_data, columns=['numTrials', 'hour'])
scaled_df.head(3)
```

```python
df[['numTrials', 'hour']] = scaled_df[['numTrials', 'hour']]
df
```

```python
unprocessed_features = ['uid', 'labname']
```

```python
encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(df[unprocessed_features])
print(one_hot_encoded.shape)

one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(unprocessed_features),  index=df.index)

df_encoded = pd.concat([df, one_hot_df], axis=1)

df_encoded = df_encoded.drop(unprocessed_features, axis=1)
df_encoded.head()
```

```python
df_encoded.to_csv('../data/dayofweek.csv')
```

```python
most_popular_class = df['dayofweek'].mode()[0]
most_popular_class
```

```python
df['naive'] = most_popular_class
accuracy_naive = accuracy_score(df['dayofweek'], df['naive'])
print(f"Accuracy naive: {accuracy_naive:.4f}")
```

## 2. Algorithms

### a. Logreg

1. Train logistic regression, for the baseline model use `random_state=21`, `fit_intercept=False`.
2. Calculate the accuracy.
3. Write a function that draws the plot (`barh`) taking coefficients of any trained models, names of the features and the number of `top-n` most important features to display.
4. Draw a plot (`barh`) for the baseline model with top-10 most important features (absolute value) for the trained model.
5. Remember that it is a multiclass classification and `coef_` returns a matrix, to calculate importance for a feature you need to sum all the individual feature importances for all the target values.

```python
X = df_encoded.drop(columns=['dayofweek'])
y = df_encoded['dayofweek']
```

```python
log_reg = LogisticRegression(random_state=21, fit_intercept=False)
log_reg.fit(X, y)
```

```python
y_pred = log_reg.predict(X)
accuracy = accuracy_score(y, y_pred)
print(f'Accuracy: {accuracy:.4f}')
```

```python
def plot_feature_importance(coef, feature_names, top_n=10):

    importance = np.sum(np.abs(coef), axis=0)

    top_indices = np.argsort(importance)[-top_n:]

    top_features = feature_names[top_indices]
    top_importance = importance[top_indices]

    plt.figure(figsize=(10, 6))
    plt.barh(top_features, top_importance, color='skyblue')
    plt.xlabel('Importance')
    plt.title(f'Top {top_n} Most Important Features')
    plt.gca().invert_yaxis()
    plt.show()

plot_feature_importance(log_reg.coef_, X.columns, top_n=10)
```

```python
def plot_feature_importance(estimators, feature_names, top_n=10):

    importance = np.sum(np.abs(np.vstack([est.coef_ for est in estimators])), axis=0)

    top_indices = np.argsort(importance)[-top_n:]

    top_features = feature_names[top_indices]
    top_importance = importance[top_indices]

    plt.figure(figsize=(10, 6))
    plt.barh(top_features, top_importance, color='skyblue')
    plt.xlabel('Importance')
    plt.title(f'Top {top_n} Most Important Features (Linear Kernel)')
    plt.gca().invert_yaxis()
    plt.show()

plot_feature_importance(svc.estimators_, X.columns, top_n=10)
```
